```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

# Set the figure size 
plt.rcParams['figure.figsize'] = (20,8)
sns.set()
```

```python
df = pd.read_csv('Data/flights_remainder.csv')
```

```python
df['lag1'] = df['remainder'].shift(1)
```

```python
df['lag2'] = df['lag1'].shift(1)
```

```python
df['lag3'] = df['lag2'].shift(1)
```

```python

sns.scatterplot(x='remainder', y='lag1',
            data=df)
```

```python
sns.scatterplot(x='remainder', y='lag2',
            data=df)
```

```python
sns.scatterplot(x='remainder', y='lag3',
            data=df)
```
